```python
import pandas as pd

bank = pd.read_csv('bankloans.csv')

bank.head()
```

```python
bank = bank.round()
bank = bank.dropna()
bank.tail()
```

```python
import plotly.express as px

fig = px.scatter_matrix(bank, dimensions=["employ", "address", "income", "debtinc","creddebt","othdebt","default"], color="default",
                        title='Variable affecting loans')
fig.show()
```

```python
import cufflinks as cf
import plotly.offline
cf.go_offline()
cf.set_config_file(offline=False)

bank.corr().iplot(kind='heatmap',colorscale='reds',title="Credit Risk Correlation")
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split

X = bank[['age','ed','employ','address','income','debtinc','creddebt','othdebt']]
y = bank['default']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
clf = LinearDiscriminantAnalysis()
clf.fit(X, y)
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions1 = logmodel.predict(X_test)
```

```python
print(classification_report(y_test,predictions1))
```
